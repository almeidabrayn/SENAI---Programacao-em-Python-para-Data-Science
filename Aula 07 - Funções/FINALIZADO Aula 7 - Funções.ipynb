```python
def taboada(numero):

  # Função que imprime a taboada para o número fornecido.

  print(f"Taboada do {numero}:")
  for i in range(1, 11):  # De 1 a 10
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

Exercício 5

Você está desenvolvendo um sistema para calcular o salário líquido de um funcionário, levando em consideração o salário base, um bônus anual e descontos de impostos e seguridade social.

As regras são as seguintes:

 - Salário Base: O salário base mensal do funcionário.
 - Bônus Anual: Um valor adicional recebido anualmente.
 - Imposto: 12% do salário base.
 - Seguridade Social: 8% do salário base.
 - O bônus é distribuído igualmente ao longo dos 12 meses do ano e adicionado ao salário base mensal.

Crie uma função que receba o salário base, o bônus anual, e retorne o salário líquido mensal do funcionário após aplicar os impostos e descontos.

image.png

```python
# Definir a função com os parâmetros slaario base e bonus
def salario_liquido(salario_base, bonus_anual):
  salario_total = salario_base - ((salario_base * 0.12) + (salario_base *0.08))
  salario_bonus = salario_total + (bonus_anual / 12)
  return salario_bonus
```

```python
salario_base = float(input("Digite o salário base mensal do funcionário: R$"))
bonus_anual = float(input("Digite o bônus anual do funcinário: R$"))

print(f"""
O salario base do funcinário é: {salario_base}
O bônus anual do funcionário é: {bonus_anual}
O salário líquido mensal do funcinário é: {round(salario_liquido(salario_base, bonus_anual),2)}
      """)
```

```python
# Definir a função com os parâmetros slaario base e bonus
def calcular_salario(salario_base, bonus):
  # Calcular o imposto e o INSS
  imposto = salario_base * 0.12
  inss = salario_base * 0.08
  salario_resultante = salario_base - imposto - inss

  # Calcular o salario resultando mais o bonus
  salario_liquido = salario_resultante + (bonus / 12)

  #retorno salario liquido
  return salario_liquido

salario_base = float(input("Digite o salário base mensal do funcionário: R$"))
bonus = float(input("Digite o bônus anual do funcinário: R$"))

# Imprimir o resultado
print(f"""
O salario base do funcinário é: {salario_base}
O bônus anual do funcionário é: {bonus}
O salário líquido mensal do funcinário é: {round(calcular_salario(salario_base, bonus),2)}
      """)
```
