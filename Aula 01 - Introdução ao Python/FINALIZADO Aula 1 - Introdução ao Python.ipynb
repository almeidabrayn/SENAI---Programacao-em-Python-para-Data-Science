## Exemplos

Declarar 3 variaveis cada uma com um número inteiro, e calcular a média dos valores, e imprimir o resultado.

```python
# Declarando as variáveis


''' Poderia ser usado assim também:
a = 10
b = 44
c = 60
'''
# var1, var2, var3

a, b, c = 10, 44, 60

# Calcular a média dos valores
media = (a+b+c)/3

# Imprimir os resultados

# Após media os :.2f é quantas casas após o resultado que será impresso;
print (f'A media é: {media:.2f}')

# round(media, 2) é outra forma, sendo a variavel e quantas casas após
print (f'A media é: {round(media, 2)}')

# Se o número não for quebrado, ele só imprimirá uma casa após o 0
```

Declarar uma variavel com o valor de um item, calcular um desconto de 20% e imprimir na tela o valor do desconto.

```python
# Declarar uma variavel com o valor do item, Variavel com 2 palavras utilizar "_"
valor_item = 505

# Calcular o desconto
desconto = valor_item * 0.22
apos_desconto = valor_item - desconto

# Imprimir o resultado

print(f'O valor do item é: {valor_item}')
print(f'O valor do desconto é: {round(desconto, 2)}')
print(f'O valor do item após o desconto é: {round(apos_desconto, 2)}')
```

3) Crie um programa que calcule o volume de uma esfera a partir de um raio.

```python
# Importar Biblioteca
import math as m

# Valores
'''diametro = float(input("Digite o diâmetro da esfera: "));
r = (diametro / 2);'''

r = float(input("Digite o valor do raio da esfera: (Basta dividir o diâmetro por 2) "));

# Calculo sem π:
volume_sem_pi = (4/3) * r**3;

# Calculos com π:
volume_com_pi = (4/3) * m.pi * r**3;

# Imprimir Valores

print(f'''
valor do raio da esfera é de: {r}
Valor do volume da esfera a partir de um raio é de apróximadamente: {round(volume_com_pi, 2)} unidades cúbicas
Valor do volume da esfera a partir de um raio em PI é de: {round(volume_sem_pi)}π unidades cúbicas''');
```

4) Criar um algoritimo que calcule um reajuste de 15% no valor de um item que será pago a prazo e depois calcule 12 parcelas desse mesmo valor e imprima:

Valor á vista:

Valor com Juros:

Parcelamento em 12x:

```python
# Valores
v_vista = float(input("Digite o valor a do produto: R$"));
parcela = int(input("Digite em quantas vezes você vai parcerlar com 15% de juros: "))

# Calculos
juros = (v_vista * 0.15);
v_juros = (v_vista + juros);
v_parcelado = (v_juros / parcela);

# Imprimir
print(f'''
O valor a vista é de: R${v_vista}
O valor do juros é de: R${juros}
O valor com juros é de: R${v_juros}
O valor parcelado em {parcela}x é de: R${round(v_parcelado, 2)}''')
```

```python
# Feito pelo Professor

# Valores
v_vista = float(input("Digite o valor a do produto: R$"));
reajuste = 1.15
num_parcela = int(input("Digite em quantas vezes você vai parcerlar com 15% de juros: "))

# Calcular os valores e prazos das parcelas
v_prazo = v_vista * reajuste
parcelas = v_prazo / num_parcela

# Imprimir os resultados
print(f'''

Valor a Vista: {v_vista}
Valor a prazo: {v_prazo}
Parcelamento em {num_parcela}x: {parcelas}
      ''')
```
