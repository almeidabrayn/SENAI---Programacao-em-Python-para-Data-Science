ELIF - Usado para mais de duas opções de resultados

```python
# Se a nota for igual ou menor que 4:
  # Imprimir a palavra ruim

# Se a nota for maior do que 4 ou menor do que 7:
  # Imprimir a palavra bom

# Se não:
  # Imprimir a palavra ótimo

nota = float(input('Valor a nota: '))

if nota <= 4:
  print('ruim')
elif 4 < nota <= 7: # outro jeito: elif nota <=7
  print('bom')
else :
  print('otimo')
```

O sistema de uma empresa categoriza seus funcionários por tempo de empresa.
São 4 classes que variam de acordo com o tempo de experiencia que o funcionário tem de empresa.



*   Treinee: até 2 anos
*   Junior:2 a 4 anos
*   Pleno: 4 a 8 anos
*   Sênior:Acima de 8



```python
# Colheta de tempo de empresa do usuário

tempo = int(input("Digite o tempo em anos de empresa: "));

# SE o tempo for menor que 2 anos
if tempo < 2 :
  # Imprimir Treinee
  print("Treinee");

# SE o tempo for menor que 4 anos
elif tempo <= 4 :
  # Imprimir Junior
  print("Junior");

# SE o tempo for menor que 8 anos
elif tempo <= 8 :
  # Imprimir Pleno
  print("Pleno")

# SE NÃO
else :
  # Imprimir Sênior
  print("Sênior")
```

### Tratamento de Exceções

Try, Except e Finally

https://docs.python.org/pt-br/3/tutorial/errors.html


Problema

```python
numero = int(input('Digite um numero inteiro: '))
if numero >= 20:
    print('O numero é maior ou igual a 20')
else:
    print('O numero é menor que 20')
```

Tratamento

```python
try:
  numero = int(input('Digite um numero inteiro: '))
  if numero >= 20:
    print('O numero é maior ou igual a 20')
  else:
    print('O numero é menor que 20')
except:
  print(' valor digitado não é um numero inteiro!')
```

Finally

```python
try:
    # Código que pode gerar um erro
    numero = int(input("Digite um número: "))
    print(f"Você digitou o número {numero}")
except:
    # O que fazer se der um erro específico (ex: valor inválido)
    print("Você não digitou um número válido!")
finally:
    # Código que sempre será executado (opcional)
    print("Fim do programa.")
```

### OBSERVAÇÃO:

Existem varios tipos de exceção e você pode trata-los individualmente se preferir, segue alguns exemplos.

```python
# ValueError
try:
    numero = int('Teste')
except ValueError:
    print('Valor inválido!')
```

```python
# TypeError
try:
    resultado = "text" + 123
except TypeError:
    print("Erro de tipo!")
```

```python
# ZeroDivisionError
try:
    resultado = 10 / 0
except ZeroDivisionError:
    print("Divisão por zero não é permitida!")
```

```python
# Captura o erro genérico
try:
    resultado = 10 / 0
except Exception as e:  # Captura o erro genérico
    print(f"Erro: {e}")
```

```python
try :
# Colhe as informações
  v_faturamento = float (input("Digite o valor de faturamento em vendas: R$"));
  salario_fixo = float(input("Digite o valor do salário fixo: R$"));

# Calculos:
  if v_faturamento <= 30000 :
    comissao = 3 / 100;
  elif v_faturamento <= 60000 :
    comissao = 5 / 100;
  else :
    comissao = 7 / 100;

# Outros Calculos
  salario_total = (comissao * v_faturamento) + salario_fixo
  comissao_total = comissao * v_faturamento

  print(f"""
  Comissão total do mês: R${round(comissao_total, 2)}
  Salário total do vendedor: R${round(salario_total,2)}
        """)
  if salario_total >= 15000 :
    print("Parabéns! Você teve um excelente mês!")
  else :
    print("Que pena!, Você não teve um excelente mês!")

except Exception as e :
  print(f"Erro: {e}");
```
