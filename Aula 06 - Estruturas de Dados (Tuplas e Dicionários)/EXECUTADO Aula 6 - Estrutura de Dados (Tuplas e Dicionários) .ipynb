## Tuplas

Tuplas são coleções ordenadas e imutáveis. Permitem armazenar múltiplos itens em uma única variável. Tuplas são definidas usando parênteses ().


###Criar

```python
tupla1 = (1, 2, 3)
tupla1
```

```python
tupla_mista = (1, "a", True)
tupla_mista
```

###Acessar

```python
tupla = (10, 20, 30, 40, 50)
```

```python
tupla[2]
```

###Imutabilidade

```python
lista = [1, 2, 3]

lista[1] = 4

lista
```

```python
tupla = (1, 2, 5)
tupla[0] = 10  # Isso gera um erro porque tuplas são imutáveis
```

###Funções em Tuplas

```python
tupla = (1, 2, 3, 2, 4, 2)
```

Contar itens da tupla

```python
len(tupla)
```

Contagem de ocorrências

```python
# Contar a ocorrência de um valor
tupla.count(2)
```

Encontrar índice

```python
# Encontrar o índice de um valor
tupla.index(3)  # 2
```

### Transformar uma lista em tupla


```python
lista = [1, 2, 3]

tupla = tuple(lista)

tupla
```

```python
tupla = (2, 3)

lista = list(tupla)

lista
```

##Dicionários

###Criação de Dicionários

```python
# Criando um dicionário vazio
dicionario_vazio = {}
```

```python
# Criando um dicionário com alguns pares chave-valor
pessoa = {
    "nome": "Guilherme",
    "idade": 33,
    "cidade": "Santo André"
}
```

###Acessando Valores


```python
print(pessoa["nome"])
print(pessoa["idade"])
```

###Modificando Valores

```python
pessoa["idade"] = 26
print(pessoa["idade"])
```

###Adicionando e Removendo Itens

```python
pessoa["profissao"] = "Instrutor"
print(pessoa)
```

```python
del pessoa["cidade"]
print(pessoa)
```

###Percorrendo Dicionários

```python
for chave in pessoa:
    print(chave)
```

```python
for valor in pessoa.values():
    print(valor)
```

```python
for chave, valor in pessoa.items():
    print(f"{chave}: {valor}")
```

###Update

```python
pessoa.update({"cidade": "São Paulo", "idade": 27})
pessoa
```

###Aninhamento de Dicionários

```python
aluno = {                       # Dicionário principal
    "nome": "João",
    "idade": 22,
    "curso": {                  # Dicionário aninhado
        "nome": "Engenharia",
        "duracao": 5
    }
}
```

Exercício 1

Crie um programa em Python que peça ao usuário para inserir o título e o autor de três livros. O programa deve armazenar essas informações em uma tupla para cada livro, e depois armazenar todas as tuplas em uma lista. Em seguida, o programa deve exibir a lista de livros com seus respectivos autores.

Requisitos:
- Solicite ao usuário que insira o título e o autor de três livros.
- Armazene essas informações em uma tupla para cada livro.
- Armazene todas as tuplas em uma lista.
- Exiba a lista de livros com seus respectivos autores.

image.png

```python
# Criar a lista vazia
livros = []

# Laço que se repete 3x
for i in range(3):
    # Nome do livro
    livro = input(f'Digite o nome do livro {i + 1}: ')
    # Autor
    autor = input(f'Digite o nome do autor do livro {livro}: ')
    # Criar a tupla
    tupla = (livro, autor)
    # Inserir a tupla na lista
    livros.append(tupla)

# Imprimir os valores da lista
print('\nLista de livros e autores: ')
for i in livros:
  print(f'Titulo: {i[0]}, Autor: {i[1]}')
```

Exercício 2

Crie um programa em Python que peça ao usuário para inserir o nome, a quantidade e o preço de três produtos. O programa deve armazenar essas informações em uma tupla para cada produto, e depois armazenar todas as tuplas em uma lista. Em seguida, o programa deve exibir a lista de produtos com suas quantidades e preços, calcular e exibir o valor total do estoque.
Requisitos:
- Solicite ao usuário que insira o nome, a quantidade e o preço de três produtos.
- Armazene essas informações em uma tupla para cada produto.
- Armazene todas as tuplas em uma lista.
- Exiba a lista de produtos com suas quantidades e preços.
- Calcule e exiba o valor total do estoque.

image.png

```python
# Criar a lista vazia
produtos = []

# Laço para colher os itens
for i in range(3):
    # Nome
    produto = input(f'Digite o nome do produto {i + 1}: ')
    # Quantidade
    quantidade = int(input(f'Digite a quantidade do produto {produto}: '))
    # Preço
    preco = float(input(f'Digite o preço do produto {produto}: '))
    # Criar a tupla
    tupla = (produto, quantidade, preco)
    # Inserir a tupla na lista
    produtos.append(tupla)

# Imprimir os valores da lista
print('\nLista de produtos')
for i in produtos:
  print(f'Produto: {i[0]}, Quantidade: {i[1]}, Preço: {i[2]}')

# Calcular o valor total do estoque
valor_estoque = 0
for i in produtos:
  valor_estoque += i[1] * i[2]
# Imprimir o valor total do estoque
print(f'\nValor tota do estoque: {valor_estoque}')
```

Exercício 3

Crie um programa para gerenciar o estoque de produtos de uma loja. O programa deve adicionar novos produtos ao estoque com dados fornecidos pelo usuário, e depois exibir um relatório com as informações de todos os produtos.


image.png

```python
# Criar o dicionário vazio
produtos = {}

# Criar um laço infinito para colher a resposta do usuario
while True:
    # Colher a resposta
    resposta = input('Deseja incluir um produto: ').lower()
    # SE a resposta for igual a sim
    if resposta == 'sim':
        # Colher codigo
        codigo = int(input('Digite o código do produto: '))
        # nome
        nome = input('Digte o nome do produto: ')
        # preço
        preco = float(input('Digite o preço do produto: '))
        # quantidade
        quantidade =int(input('Digite a quantidade do produto: '))
        # Criar a chave, e o dicionario como valor
        produtos[codigo] = {'nome': nome, 'preco': preco, 'quantidade': quantidade}
    # SE a resposta for igual a não
    elif resposta == 'não':
        # Interromper o laço
        break
    # SE NÃO
    else:
        # Imprimi a menssagem de entrada inválida
        print('Entrada inválida!')
        # Continuar no laço
        continue

# Imprimir o relatório
print('\nRelatório de estoque')
for i in produtos.values():
  print(f'''
  Nome: {i['nome']}
  Preço: {i['preco']}
  Quantidade: {i['quantidade']}
  ''')
```
