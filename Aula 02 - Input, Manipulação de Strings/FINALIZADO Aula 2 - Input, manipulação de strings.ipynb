2) Tirar a média de 3 notas de prova

```python
# Colher notas das 3 provas do usuário

n1 = float (input("Digite a primeira nota: "));
n2 = float (input("Digite a segunda nota: "));
n3 = float (input("Digite a terceira nota: "));

# Calcular a média das notas
media = ((n1+n2+n3)/3);

# Imprimir o resultado
if media >= 6:
  print(f'Aprovado com média: {round(media, 2)}');
else:
  print(f'Reprovado com média: {round(media, 2)}');
```

Média através da Biblioteca Statistics

https://docs.python.org/pt-br/3/library/statistics.html

```python
import statistics as sta

media_2 = sta.mean([n1, n2, n3])

print(f'O valor da média é: {round(media_2,2)}')
```

1) Crie um programa que colha um raio do usuário e calcule a área de um circulo a partir do raio informado.

Dica: Pesquisar na internet a formula usada para calcular a area através da constante Pi, e aplicar no algoritimo a função da biblioteca math que traz esse valor.

https://docs.python.org/pt-br/3/library/math.html



```python
# Importar bibliotecas
import math as m

# Colher o raio do usuário
r = float (input("Digite o valor do raio do circulo: "));

# Calcula a área do circulo a partir do raio

area = (m.pi * r ** 2);

print(f'''
O raio do circulo é: {r}
A área do circulo é: {round(area, 2)}cm²
''');
```

3) Crie um programa que peça ao usuário para inserir uma temperatura em graus Celsius e a converta para Fahrenheit e Kelvin. O programa deve exibir as duas temperaturas convertidas.

image.png

```python
# Colhe as informações da temperatura em ºC
celsius = round(float(input("Digite o valor em Celsius para a converção: ")) ,2);

# Converte Celsius para Fahrenheit e Kelvin
fahrenheit = round((celsius * 9/5) + 32 ,2);
kelvin = round(celsius + 273.15 ,2);

# Imprime os resultados

print(f"""
Valor em Celsius: {celsius}ºC
Valor em Fahrenheit: {fahrenheit}ºF
Valor em Kelvin: {kelvin} K
""");
```

4) Crie um programa que peça ao usuário para inserir o valor original de um produto e a porcentagem de desconto. O programa deve calcular e exibir o valor do desconto e o valor final do produto.


image.png

```python
# Colhe os valores do produto
produto_original = int (input("Digite o valor original do produto: R$"));
desconto = int (input("Digite o valor do desconto (em %): "));

# Calcula os valores
v_desconto = desconto / 100;
p_desconto_valor = (produto_original * v_desconto);
p_desconto = round(produto_original - (produto_original * v_desconto));

# Imprime os valores
print(f"""
O valor original do produto: R${produto_original}
O valor da procentagem de desconto: {desconto}%
O valor do desconto é: R${p_desconto_valor}
O valor final do produto é: R${p_desconto}""");
```

5) Crie um programa que peça ao usuário para inserir o valor total de uma conta de restaurante, a porcentagem de gorjeta que deseja dar, e o número de pessoas para dividir a conta. O programa deve calcular e exibir o valor da gorjeta, o valor total com a gorjeta e quanto cada pessoa deve pagar.

image.png

```python
# Colhe os valores do usuário
conta = int (input("Digite o valor total da conta: "));
porcentual = int (input("Digite o valor da gorjeta (em %): ")) / 100;
pessoas = int (input("Digite o número de pessoas para dividir a conta: "));

# Calcula os valores;
gorjeta = conta * porcentual;
v_conta_com_gorjeta = conta + gorjeta;
v_pessoa = v_conta_com_gorjeta / pessoas;

# Imprimir valores
print(f'''
Valor total da conta: R${conta}
Valor da gorjeta (em %): {porcentual * 100}%
Número de pessoas para dividir a conta: {pessoas}

O valor da gorjeta é: R${gorjeta}
O valor total com a gorjeta é: R${v_conta_com_gorjeta}
Cada pessoa deve pagar: R${v_pessoa}''');
```
