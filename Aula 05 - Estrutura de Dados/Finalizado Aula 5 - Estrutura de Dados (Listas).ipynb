###Exemplo

Exemplo 1 - Lista de tarefas

```python
# Inicializando a lista de tarefas vazia
tarefas = []

# Adicionando tarefas à lista
tarefas.append("Comprar pão")
tarefas.append("Ir para aula de python")
tarefas.append("Ler um livro")

# Imprimindo a lista de tarefas
print("Lista de Tarefas Inicial:")
for tarefa in tarefas:
    print(tarefa)

# Modificando uma tarefa
tarefas[1] = "Estudar para a prova de Python"

# Adicionando mais uma tarefa
tarefas.append("Fazer exercícios físicos")

# Removendo uma tarefa
tarefas.remove("Comprar pão")

# Imprimindo a lista de tarefas atualizada
print("\nLista de Tarefas Atualizada:")
for tarefa in tarefas:
    print(tarefa)

# Exibindo o tamanho da lista de tarefas
print(f"\nNúmero total de tarefas: {len(tarefas)}")
```

Exemplo 2 - Lista de Carros

```python
# Inicializando a lista de carros
carros = []

# Solicitando ao usuário quantos modelos de carros ele deseja inserir
numero_de_carros = int(input("Quantos modelos de carros você deseja inserir? "))

# Laço for para inserir os nomes dos carros na lista
for i in range(numero_de_carros):
    modelo = input(f"Digite o nome do modelo de carro {i + 1}: ")
    carros.append(modelo)

# Imprimindo a lista de carros inseridos
print("\nLista de modelos de carros inseridos:")
for carro in carros:
    print(carro)
```

Exemplo 3 - Pontos em programa de relacionamento de clientes

```python
# Inicializando as listas vazias
clientes = []
pontos = []

# Solicitando ao usuário para inserir os nomes dos clientes e suas quantidades de pontos
for i in range(2):
    cliente = input(f"Digite o nome do cliente {i + 1}: ")
    quantidade_pontos = int(input(f"Digite a quantidade de pontos de {cliente}: "))
    clientes.append(cliente)
    pontos.append(quantidade_pontos)

# Imprimindo a lista de clientes com suas quantidades de pontos
print("\nLista de clientes e suas quantidades de pontos:")
for i in range(2):
    print(f"Cliente: {clientes[i]}, Pontos: {pontos[i]}")
```

Exercício 2

Crie um programa em Python que solicite ao usuário para inserir os nomes de três países e armazene esses nomes em uma lista. Em seguida, peça ao usuário para inserir o nome de um país adicional e a posição em que ele deseja inserir este país na lista. Por fim, exiba a lista completa de países.

image.png

```python
# Criar a lista de países vazia
paises = []
ordem = ['primeiro', 'segundo', 'terceiro']

# Laço que se repetirá por 3x
for i in ordem:
    # Colher o nome do país
    pais = input(f'Digite o nome do {i} país: ')
    # Inserir esse país na lista
    paises.append(pais)

# Imprimir a lista
print(f'\nLista de países atual: {paises}')

# Colher o nome de um país adicional
pais_adicional = input('Digite o nome de um país adicional: ')

# Colher a posição em que se deseja inserir esse país
while True:
  posicao = int(input('\nDigite a posição que deseja inserir (0, 1, 2 e 3): '))
  if posicao < 0 or posicao > 4:
    print('Entrada inválida, digite uma posição entre 0 e 3')
  else:
    break
# Inserir o país na lista
paises.insert(posicao, pais_adicional)

# Imprimir a lista atualizada
print(f'\nlista de países atualizada: {paises}')
```

Exercício 3

Crie um programa em Python que solicite ao usuário para inserir os nomes de cinco objetos e armazene em uma lista. Em seguida, peça ao usuário para inserir o nome de um objeto que ele deseja remover da lista. Por fim, remova o objeto especificado da lista e exiba a lista atualizada.

image.png

```python
# Criar a lista vazia
objetos = []

# Laço que se repetirá por 5x
for i in range(5):
    # colher  o nome do objeto
    objeto = input(f'Digite o nome do {i + 1}° objeto: ')
    # Inserir o objeto na lista
    objetos.append(objeto)

# Imprimir a lista
print(f'\nLista de objtos atual: {objetos}')

# Colher o nome do objeto que ele deseja retirar
while True:
  objeto_retirar = input('Digite o nome do objeto que deseja retirar: ')
  if objeto_retirar not in objetos:
    print('Este objeto não está na lista!')
  else:
    break

# Retirar o objeto da lista
objetos.remove(objeto_retirar)

# Imprimir a lista atualizada
print(f'\nLista atualizada: {objetos}')
```

Exercício 4

Crie um programa em Python que simule o controle de desempenho de três atletas de alto rendimento. O programa deve permitir que o usuário insira o nome dos atletas e suas respectivas pontuações em uma competição. Em seguida, o programa deve solicitar ao usuário para atualizar a pontuação de um atleta específico e, por fim, exibir a lista atualizada de atletas com suas pontuações.



image.png



```python
# Criando as listas de atletas e pontuações
atletas = []
pontuacoes = []

# Laço que se repetirá por 3x
for i in range(3):
    # Colher o nome do atleta
    atleta = input(f'Digite o nome do {i + 1}° atleta: ')
    # Colher a pontuação
    pontuacao = int(input(f'Digite a pontuação do atleta {atleta} '))
    # Inserir os valores nas listas
    atletas.append(atleta)
    pontuacoes.append(pontuacao)

# Imprimir a lista atual
print('\nLista de atletas atual: ')
for i in range(len(atletas)):
  print(f'Atleta: {atletas[i]}, pontuação: {pontuacoes[i]}')

# Colher o nome do atleta que deseja atualizar a pontuação
atleta_modificar = input('Digite o nome do atleta que deseja modificar a pontuação: ')
# Colher a nova pontuação
pontuacao_modificar = int(input(f'Digite o valor da nova pontuação do atleta {atleta_modificar}: '))
# Descobrir o indice da pontuação a ser modificada
indice = atletas.index(atleta_modificar)
# Atualizar a pontuação
pontuacoes[indice] = pontuacao_modificar


# Imprimir a lista atualizada
print('\nLista de atletas atualizada: ')
for i in range(len(atletas)):
  print(f'Atleta: {atletas[i]}, pontuação: {pontuacoes[i]}')
```
