### Alterando as dimensões da figura e do gráfico

###Intervalo de Data

```python
# Importar biblioteca Date Time para Selecionar um intervalo de data Específico

from datetime import datetime
figura1 = plt.figure(figsize=(10,10))
eixo1 = figura1.add_axes([0,0,2,1])
eixo1.plot(usim5.Date, usim5.Close, color = 'springgreen', linewidth=5)
eixo1.set_title('Usiminas - USIM5', fontsize=50, pad=30)
eixo1.set_ylabel('Valor por ação em R$', fontsize=30)
eixo1.set_xlabel('Data', fontsize=30)
eixo1.legend(['Valor da ação em R$'], loc='upper right', fontsize=30)
eixo1.set_facecolor('#008080')
eixo1.grid(False)

# Limitar o X em um valor de Start / End e a mesma coisa com o Y
eixo1.set_xlim(datetime(2014,1,1), datetime(2016,1,1))
```

###Trabalhando com dois eixos.

```python
figura1 = plt.figure(figsize=(10,10))
eixo1 = figura1.add_axes([0,0,2,1])

# movimenta o gráfico menor
# Eixo X / Y | Largura do X | Largura do Y
eixo2 = figura1.add_axes([1,0.5,0.8,0.3])


eixo1.plot(usim5.Date, usim5.Close, color = 'springgreen')
eixo1.set_title('Usiminas - USIM5', fontsize=50, pad=30)
eixo1.set_ylabel('Valor por ação em R$', fontsize=30)
eixo1.set_xlabel('Data', fontsize=30)
eixo1.legend(['Valor da ação em R$'], loc='upper right', fontsize=30)
eixo1.set_facecolor('#008080')
eixo1.grid(False)


eixo2.plot(usim5.Date, usim5.Close, color = 'springgreen')
eixo2.set_title('2014 - 2016')
eixo2.set_xlabel('Data', fontsize=30)
eixo2.set_xlim(datetime(2014,1,1), datetime(2016,1,1))
eixo2.set_facecolor('#5F9EA0')
```

```python
figura1 = plt.figure(figsize=(10,10))
eixo1 = figura1.add_axes([0,0,2,1])
eixo2 = figura1.add_axes([1.05,0.55,0.9,0.4])


eixo1.plot(usim5.Date, usim5.Close, color = 'springgreen')
eixo1.set_title('Usiminas - USIM5', fontsize=50, pad=30)
eixo1.set_ylabel('Valor por ação em R$', fontsize=30)
eixo1.set_xlabel('Data', fontsize=30)
eixo1.legend(['Valor da ação em R$'], loc='upper left', fontsize=30)
eixo1.set_facecolor('#008080')
eixo1.grid(False)


eixo2.plot(usim5.Date, usim5.Close, color = 'springgreen')
eixo2.set_title('2014 - 2016', fontsize=20, color='white')
eixo2.set_xlim(datetime(2014,1,1), datetime(2016,1,1))
eixo2.set_facecolor('#5F9EA0')
```

Enfase para a área do grafico menor.

```python
figura1 = plt.figure(figsize=(10,10))
eixo1 = figura1.add_axes([0,0,2,1])
eixo2 = figura1.add_axes([1.05,0.55,0.9,0.4])

# Variavel Ano | Mês | Dia
esquerda = usim5.Date < datetime(2014,1,1)
direita =  usim5.Date > datetime(2016,1,1)

# Cria o gráfico inteiro com a linha vermelha
eixo1.plot(usim5.Date, usim5.Close, color = 'red')

# Pede para criar uma linha verde, só que tudo antes de 2014
eixo1.plot(usim5[esquerda]['Date'], usim5[esquerda]['Close'], color = 'springgreen')

# Pinta de verde só o de 2016 para frente
eixo1.plot(usim5[direita]['Date'], usim5[direita]['Close'], color = 'springgreen')


eixo1.set_title('Usiminas - USIM5', fontsize=50, pad=30)
eixo1.set_ylabel('Valor por ação em R$', fontsize=30)
eixo1.set_xlabel('Data', fontsize=30)
eixo1.legend(['Valor da ação em R$'], loc='upper left', fontsize=30)
eixo1.set_facecolor('#008080')
eixo1.grid(False)

# Apenas para destacar o gráfico acima foi alterado
eixo2.plot(usim5.Date, usim5.Close, color = 'red')
eixo2.set_title('2014 - 2016', fontsize=20, color='white')
eixo2.set_xlim(datetime(2014,1,1), datetime(2016,1,1))
eixo2.set_facecolor('#5F9EA0')

# Comparando, quem comprou a 40 reais o risco era muito alto
```

1) Importar a base de dados e criar um grafico de linha com o histórico das ações, e um grafico de enfase para o período da pandemia da covid-19.


image.png



```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

url = '/content/drive/MyDrive/Programação em Python para Data Science/VALE3.SA.csv'
vale3 = pd.read_csv(url)
vale3.head()
```

```python
# Transformar em Data está como objeto
vale3.Date = pd.to_datetime(vale3.Date)

vale3.info()
```

```python
# Criando a figura e o eixo
figura1 = plt.figure(figsize=(15,8))
eixo1 = figura1.add_axes([0,0,2,1])
eixo2 = figura1.add_axes([0.8,0.565,0.65,0.3])

# Muda a cor da fugura - Dentro
eixo1.set_facecolor('#78948c')

# Cor do Fundo da figura - Externa
figura1.set_facecolor('#486a64')

# Criar as variaveis que determinam o antes e o depois da covid Ano | Mês | Dia
esq_covid = vale3.Date < datetime (2020,5,1)
dir_covid =  vale3.Date > datetime(2023,5,1)

# Cria o gráfico inteiro com a linha vermelha / Plotar o eixo 1
eixo1.plot(vale3.Date, vale3.Close, color = 'red')

# Pede para criar uma linha verde, só que tudo antes de 2014
eixo1.plot(vale3[esq_covid]['Date'], vale3[esq_covid]['Close'], color = 'black')
eixo1.plot(vale3[dir_covid]['Date'], vale3[dir_covid]['Close'], color = 'black')

eixo1.set_title('Vale - VALE3', fontsize=50, pad=30, color = '#78948c')
eixo1.set_ylabel('Valor por ação em R$', fontsize=30, color = '#78948c')
eixo1.set_xlabel('Data', fontsize=30, color = '#78948c')
eixo1.set_xlim(datetime(2000,1,1), datetime(2023,6,15))
eixo1.legend(['Valor da ação em R$'], loc='lower right', fontsize=30, frameon = False, labelcolor='#16413a')
eixo1.grid(True)

# Plotar o eixo 2 / grafico 2
eixo2.plot(vale3.Date, vale3.Close, color = 'red')
eixo2.set_title('COVID', fontsize=20, color='#16413a')
eixo2.set_xlim(datetime(2020,5,1), datetime(2023,5,1))
eixo2.set_facecolor('#203c3c')
eixo2.grid(True)
```

2) Plotar um gráfico de barras com o valor médio de fechamento das ações, agrupado por ano.


image.png

```python
# Volume de Fechamento
volume_por_fechamento = vale3[['Date',
                               'Close']].groupby(pd.Grouper(key='Date',
                                                            freq='YE')).mean("close")
```

```python
fig3 = plt.figure(figsize=(10,4))
eixo4 = fig3.add_axes([0,0,1,1])


# Cor do Fundo da figura - Externa
fig3.set_facecolor('#78948c')


eixo4.bar(volume_por_fechamento.index, volume_por_fechamento.Close, width=300, edgecolor = "#71b584", color = '#78948c')
eixo4.set_title('volume negociado por ano')
eixo4.set_ylabel('Preço das ações em TRI R$')
eixo4.set_xlim(datetime(2000,7,1), datetime(2023,6,30))
eixo4.set_xlabel('ANO')
eixo4.set_facecolor('#486a64')
```
