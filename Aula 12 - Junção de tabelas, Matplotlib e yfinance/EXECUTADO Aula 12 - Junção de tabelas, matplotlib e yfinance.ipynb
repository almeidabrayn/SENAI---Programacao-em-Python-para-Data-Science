## Junção de Tabelas

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as datetime

url1_vale = '/content/drive/MyDrive/Programação em Python para Data Science/VALE3.SA.csv'

url2_usim = '/content/drive/MyDrive/Programação em Python para Data Science/USIM5.SA.csv'

vale3 = pd.read_csv(url1_vale)

usim5 = pd.read_csv(url2_usim)


vale3.Date = pd.to_datetime(vale3.Date)


usim5.Date = pd.to_datetime(usim5.Date)
```

###Função Concat

```python
pd.concat([vale3, usim5])
```

```python
reset_indice = pd.concat([vale3, usim5], ignore_index=True)
```

```python
reset_indice
```

### Merge

```python
# suffixess ele altera o parametro que apareceria de high value x para high vale Vale e High Usiminas - Apenas para colunas em comum
df_juntos = pd.merge(vale3, usim5, on='Date', suffixes=['_vale', '_usim'])

df_juntos
```

## MatplotLib

###Comparação de valorização %

```python
fig1 = plt.figure(figsize=(15,8))
eixo1 = fig1.add_axes([0,0,2,1])

eixo1.plot(df_juntos.Date, df_juntos.Close_vale, color='green')
eixo1.plot(df_juntos.Date, df_juntos.Close_usim, color='black')
```

1) Colher os dados do mesmo período de duas empresas listadas na bolsa de valores que possuem o mesmo seguimento, e plotar um grafico de comparação entre as duas. - Mesma bolsa de valores e Mesma Moeda para comparar

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as datetime
import yfinance as y

# Periodos: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max


nintendo = y.Ticker('7974.T')
df_nintendo = nintendo.history('10y')


sony = y.Ticker('6758.T')
df_sony = sony.history('10y')

df_nintendo.reset_index(inplace=True)
df_sony.reset_index(inplace=True)

df_juntos = pd.merge(df_sony, df_nintendo, on='Date', suffixes=['_sony', '_nintendo'])

df_juntos.head(5)
```

```python
df_juntos['Close_nintendo'][0]
df_juntos['var%_nintendo']= df_juntos['Close_nintendo'].div(df_juntos['Close_nintendo'][0])-1
df_juntos['var%_sony']= df_juntos['Close_sony'].div(df_juntos['Close_sony'][0])-1
df_juntos.head(5)
```

```python
fig1 = plt.figure(figsize=(15,8))
eixo1 = fig1.add_axes([0,0,2,1])

eixo1.plot(df_juntos.Date, df_juntos['var%_nintendo'], color='green')
eixo1.plot(df_juntos.Date, df_juntos['var%_sony'], color='black')
```
