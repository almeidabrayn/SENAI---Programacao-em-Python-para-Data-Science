```python
# Importar os dados
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
import joblib

titanic = "/content/drive/MyDrive/Programação em Python para Data Science/Titanic/titanic.xlsx"

df_titanic = pd.read_excel(titanic)
```

```python
# Tratar os dados
df_titanic.columns
df_titanic.drop(['index','PassengerId', 'Name'], axis = 1, inplace = True)

map_sex = {'male': 1, 'female': 0}
df_titanic['Sex'] = df_titanic['Sex'].replace(map_sex)

df_titanic["Age"] = df_titanic["Age"].astype(int)

df_titanic.head()
```

```python
# Fazer a separação dos dados
x = df_titanic.drop('Survived_y', axis = 1)

y = df_titanic.Survived_y

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size= 0.10, random_state = 1)
```

```python
# Treinar

modelo = DecisionTreeClassifier(max_depth=1)
modelo.fit(x_treino, y_treino)
```

```python
# Testar
previsoes = modelo.predict(x_teste)
accuracy_score(y_teste, previsoes)
acuracia = accuracy_score(y_teste, previsoes) * 100
acuracia
```

```python
# Criar o arquivo pkl
joblib.dump(modelo, 'modelo_titanic.pkl')
print("Modelo salvo com sucesso!")
```

# Modelo Simples

```python
# Colher e prever
modelo_treinado = joblib.load('/content/modelo_titanic.pkl')

var1 = int(input('Qual a classe do passageiro? (1 a 3) '))
var2 = int(input('Qual o sexo do passageiro? (0 Feminino e 1 para Masculino) '))
var3 = int(input('Qual a idade do passageiro? '))
var4 = int(input('Tinha irmãos a bordo? (0 para Não e 1 para Sim) '))
var5 = int(input('Tinha pais a bordo? (0 para Não e 1 para Sim) '))
var6 = float(input('Qual o valor da passagem? '))

df = pd.DataFrame({
    'Pclass' : [var1],
    'Sex' : [var2],
    'Age' : [var3],
    'SibSp' : [var4],
    'Parch' : [var5],
    'Fare': [var6]
})

previsao_sobrevivente = modelo_treinado.predict(df)

if previsao_sobrevivente == 0:
  print("""
O passageiro em breve não estará entre nós""")
else:
  print("""
O passageiro viverá para ver a luz do dia novamente""")
```
